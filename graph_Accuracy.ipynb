```python
import matplotlib.pyplot as plt

# Define the data for the histogram
methods = ["2D_LSTM", "2D_LSTM_pre", "RNN", "LSTM_Teheral_mine", "MWPM"]
accuracy_scores = [0.90, 0.91, 0.88, 0.82, 0.89]  # Replace these with your actual accuracy values

# Create a list of colors, with the last one as 'orange'
colors = ['skyblue'] * (len(methods) - 1) + ['orange']

# Create the histogram
plt.figure(figsize=(10, 6))
plt.bar(methods, accuracy_scores, color=colors)
plt.xlabel("Method")
plt.ylabel("Accuracy (5 Rounds)")
plt.title("Accuracy of Different Methods over 5 Rounds (Surface_Dataset)")
plt.ylim(0.7, 1)  # Sets the y-axis from 0.5 to 1
plt.show()

```
